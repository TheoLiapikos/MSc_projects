```python
# Load dictionary structure file containing LDA analysis data
with open(path_to_lda_data_dictionary, "rb") as fp:
  lda_data = pickle.load(fp)

# Get Dates
dates = lda_data['dates']

# Load combined emerging topics list
# with open(path_to_comb_emerg_topics_list, "rb") as fp:
#   comb_emerg_topics = pickle.load(fp)
```

```python
# Lets plot topics for all 15 dates
idx = 0
```

```python
# Let work with a specific date's topics in list
date = dates[idx]
print('\n Plot topics for day: %s' %date)
idx+=1

# Get LDA analysis data for this particular date
lda_model,bow_corpus,dictionary = lda_data[date]

# Visualize the topics-keywords  - Less than 2 min
pyLDAvis.enable_notebook()
hashtags_vis = pyLDAvis.gensim.prepare(lda_model, bow_corpus, dictionary)
hashtags_vis
```
